```python
import scipy.optimize as opt 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
c=np.array([10,15])
A=np.array([[-2,-1],[-.3,-.6],[-.2,-.2]])
b=np.array([-100,-25,-10])
sol=opt.linprog(c,A,b)
sol.x
```

```python
# Definir funciones de restricción y de costo
def r1(xB):
    return (-100+1*xB)/-2
def r2(xB):
    return (-25+.6*xB)/-.3
def r3(xB):
    return (-10+0.2*xB)/-.2

# Graficar
xB=np.linspace(0,400,100)
plt.plot(xB,r1(xB),label="r1") #se etiqueta cada restricción 
plt.plot(xB,r2(xB),label="r2")
plt.plot(xB,r3(xB),label="r3")
plt.legend(loc="best") #comando para mostrar el cuadro de etiquetas 
plt.xlabel("$x_B$") # se etiqueta eje "x" como xB
plt.ylabel("$x_A$")
plt.axis([0,100,-10,100]) #se recortan los ejes 
plt.grid()
```
